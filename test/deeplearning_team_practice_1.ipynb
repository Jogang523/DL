```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 아래의 데이터에서 Yearly Amount Spent를 예측하는 모델을 머신러닝 모델과 딥러닝 neural net 모델로 각각 생성하고,
# 모델을 저장하시오
```

```python
df = pd.read_csv('C:\\Users\\M\\Downloads\\Bigdata\\Ecommerce Customers.txt')
df.head(3)
```

```python
# 머신러닝
```

```python
#타겟 티쳐 분리
y = df['Yearly Amount Spent']
X = df.drop(['Yearly Amount Spent', 'Email', 'Address', 'Avatar'], axis=1)
```

```python
#학습,테스트 데이터 분리
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#학습
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
#테스트
preds = lm.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_test, preds))
print(np.sqrt(mean_squared_error(y_test, preds)))
```

```python
#딥러닝
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#딥러닝 모델 생성
model = Sequential()
model.add(Dense(48, input_dim = 4, activation='linear'))
model.add(Dense(24, activation='linear'))
model.add(Dense(12, activation='linear'))
model.add(Dense(6, activation='linear'))
model.add(Dense(1))
```

```python
#딥러닝에선 넘파이 배열을 넣는게 더 좋다.
x_train = x_train.values.astype('float32')
x_test = x_test.values.astype('float32')
y_train = y_train.values.astype('float32')
y_test = y_test.values.astype('float32')
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
preds = model.predict(x_test)
print(r2_score(y_test, preds))
print(mean_squared_error(y_test, preds))
```

```python
# 아래의 코드에서 TARGET CLASS를 분류하는 머신러닝 모델과 딥러닝 neural network모델 중 binary classification과
# multi classification 두 방식으로 neural net 모델을 생성하고, 모델을 저장하시오
```

```python
df = pd.read_csv('C:\\Users\\M\\Downloads\\Classified Data.csv',index_col=0)
df.head(3)
```

```python
y=df['TARGET CLASS']
x=df.drop('TARGET CLASS',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfr=RandomForestClassifier()
rfr.fit(x_train,y_train)
```

```python
# binary classification
```

```python
#모델 생성(마지막 노드는 바이너리 이기 때문에 시그몬드, 노드의 수는 1 고정)
model = Sequential()
model.add(Dense(48, input_dim = 10, activation='relu'))
model.add(Dense(24, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
#binary로 로스 설정
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#학습 & 테스트
model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test))
```

```python
# multi classification
```

```python
#모델 생성, classification이기 때문에 마지막 댄스 레이아웃은 노드 2 고정, activation='softmax' 고정
model = Sequential()
model.add(Dense(48, input_dim = 10, activation='relu'))
model.add(Dense(24, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(2,activation='softmax'))
```

```python
#classification은 카테고리 형식으로 나와야 하기 떄문에 카테고리 형식으로 변환
from tensorflow.keras.utils import to_categorical

y_train_ohe = to_categorical(y_train)
y_test_ohe = to_categorical(y_test)
```

```python
#categorical로 로스 설정
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#학습 & 테스트
model.fit(x_train,y_train_ohe,epochs=100,validation_data=(x_test,y_test_ohe))
```
