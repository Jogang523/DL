```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from tensorflow.keras.layers import Dense, Embedding, Dropout, LSTM, MaxPooling1D, Conv1D, GlobalMaxPooling1D
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
import re

from keras.utils import to_categorical
```

```python
file = 'C:\\Users\\M\\Downloads\\Bigdata\\lastofus.txt'
```

```python
with open(file, 'r', encoding='euckr') as f:
  lines = f.readlines()
```

```python
len(lines)
```

```python
for line in lines:
  print(line)
```

```python
conv = [line for line in lines if re.match(r'^[A-Z].*:', line)] #    ":"   =>  
```

```python
conv
```

```python
text = ' '.join(conv)
```

```python
text
```

```python
arr = []
for line in text.split('\n'):
  arr.append(line.strip())
```

```python
arr
```

```python
# text

# 1.  
# 2. tokenizing :  ( , )
# 3. word index :  index 
# 4.  => token index 
# 5.   ...
# 6. vectorizing :
      #  vector   : CountVectorizer, TfidfVectorizer
      # (token) vector   : Embedding
```

```python
# tokenizing =>  token ...
```

```python
token = Tokenizer(lower=False, filters='.,?;:\'\"-')
token.fit_on_texts(arr)
```

```python
token.word_index
```

```python
#  => token index 
```

```python
seq = token.texts_to_sequences(arr)
seq
```

```python
#    
```

```python
seq = pad_sequences(seq, maxlen=15)  #    0     15 
seq
```

```python
seq.shape
```

```python
# x, y ..
```

```python
x = seq
y = np.vstack((seq[1:], seq[0]))
x = seq[:-1]
y = y[:-1]
```

```python
x.shape
```

```python
y.shape
```

```python
len(token.word_index)
```

```python
y[:2]
```

```python
# token index => category => one hot
```

```python
y_ohe = to_categorical(y)
```

```python
y_ohe.shape  # token => 2627 one hot vector
```

```python
# LSTM data shape : (batch_size, sequence length, input dim)
```

```python
#
x.shape
```

```python
x_ohe = to_categorical(x)
```

```python
x_ohe.shape
```

```python
model = Sequential()
model.add(LSTM(1284, input_shape=(15, 2627), return_sequences=True))
model.add(Dense(2627, activation='relu'))
model.add(Dense(2627, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_ohe,y_ohe, epochs=100, batch_size=30)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
