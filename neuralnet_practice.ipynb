```python
import pandas as pd
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
```

```python
df = pd.read_csv('C:\\Users\\M\\Downloads\\Bigdata\\sonar.csv', header=None)
df.head()
```

```python
df[60].value_counts()
```

```python
# 위의 데이터를 기반으로 neuralnetwork를 학습시켜서 M과 R을 예측하는 모델을 생성하고 accuracy를 계산하시오.
# 단, 이진분류모델과 다중분류모델을 각각 생성하시오
```

```python
#다중분류모델.
```

```python
y=df[60]
x=df.drop(60,axis=1)
```

```python
label_encoder=LabelEncoder()
y_in = label_encoder.fit_transform(y)
```

```python
from tensorflow.keras.utils import to_categorical
y_ohe = to_categorical(y_in)
```

```python
model=Sequential()
model.add(Dense(30,input_dim=60,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(2,activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y_ohe,epochs=100)
```

```python
model.evaluate(x, y_ohe)
```

```python
df = pd.read_csv('C:\\Users\\M\\Downloads\\Bigdata\\xAPI-Edu-Data.csv')
df.head()
```

```python
# 위의 데이터를 기반으로 neuralnetwork 모델을 학습시켜서 Class를 예측하는 모델을 학습시키고, 모델을 평가하시오..
# neural network 모델을 다지분류 모델을 생성하시오..
```

```python
df
```

```python
y=df['Class']
x=df.drop('Class',axis=1)
```

```python
x_ohe=pd.get_dummies(x,dtype=int)
```

```python
len(x_ohe.columns)
```

```python
label_encoder=LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
y_ohe = to_categorical(y)
```

```python
model=Sequential()
model.add(Dense(30,input_dim=72,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
x_ohe
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_ohe,y_ohe, epochs=100)
```
