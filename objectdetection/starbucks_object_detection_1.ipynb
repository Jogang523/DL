```python
import yaml

data = {
    'train': '/starbucks/data/images',
    'val': '/starbucks/data/images',
    'test':'/starbucks/data/images',
    'nc': 2,
    'names': ['letter', 'logo']
}

with open('../starbucks/custom.yaml', 'w') as f:
    yaml.dump(data, f)
```

```python
!python train.py --data ../starbucks/custom.yaml --weights yolov5m.pt

# --data ../starbucks/custom.yaml: 데이터셋 구성 정보를 담고 있는 YAML 파일의 경로를 지정합니다. 이 파일에는 학습, 검증 및 테스트 데이터셋의 경로와 클래스 목록 등이 포함되어 있습니다.
# --weights yolov5m.pt: 사전 학습된 YOLOv5 중간 크기(medium) 모델의 가중치 파일을 지정합니다. 이 가중치를 사용하여 모델을 초기화하고 추가적인 학습을 수행합니다.
```
